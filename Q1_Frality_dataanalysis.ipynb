```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("clean_data_fratality.csv")

data.sort_values(by='Age', inplace=True)
data.reset_index(drop=True, inplace=True)
```

```python
data
```

```python
x=data['Height(inches)']
Y=data['Weight(pounds)']
plt.bar(x,Y)
```

```python
sns.boxplot(x='Frailty', y='Age', data=data)
plt.title('Age by Frailty')
plt.show()
```

```python
x=data['Age']
y=data['Weight(pounds)']
plt.scatter(x, y, c='blue', marker='o', label='Data Points')
plt.xlabel('Grip')
plt.ylabel('weight')
plt.legend()
plt.grid(True)
plt.show()
```
