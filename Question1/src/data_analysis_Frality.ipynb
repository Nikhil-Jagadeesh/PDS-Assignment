```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('clean_data_fratality.csv')

data.sort_values(by='Age', inplace=True)
data.reset_index(drop=True, inplace=True)
```

```python
data
```

```python
x = data['Weight(pounds)']
Y = data['Height(inches)']
plt.scatter(x, Y, marker='o', color='red', alpha=0.5)
plt.xlabel('Height(inches)')
plt.ylabel('Weight(pounds)')
plt.title('Scatter Plot of Height vs. Weight')
plt.show()
```

```python
custom_palette = sns.color_palette("pastel")
plt.figure(figsize=(10, 6))
sns.boxplot(x='Frailty', y='Age', data=data, palette=custom_palette, showfliers=False, width=0.5)
plt.xlabel('Frailty')
plt.ylabel('Age')
plt.title('Box Plot: Age by Frailty')

plt.show()

```

```python
x = data['Age']
y = data.groupby('Age')['Weight(pounds)'].mean().values
plt.bar(x, y, color='skyblue', label='Average Weight')
plt.xlabel('Age (years)')
plt.ylabel('Weight (pounds)')
plt.title('Bar Plot of Age vs. Weight')
plt.legend()
plt.show()
```

```python
x = data['Age']

plt.plot(x, label='Age', color='blue', marker='o', linestyle='-')

# Customize labels, title, and legend
plt.xlabel('Data Points')
plt.ylabel('Age')
plt.title('Line Plot of Age')
plt.legend()
plt.grid()
plt.show()

```
